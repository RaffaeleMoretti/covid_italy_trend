```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
dati = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv")
dati
```

```python
dati['deceduti']=dati['deceduti']-dati['deceduti'].shift()
```

```python
dati.rename(columns = {'deceduti':'nuovi_deceduti'}, inplace = True)
```

```python
dati.fillna(0, inplace=True)
dati['nuovi_deceduti']=dati['nuovi_deceduti'].astype(int)
dati["nuovi_deceduti"].loc[0] = 7
pd.set_option("mode.chained_assignment", None)
dati
```

```python
datinuovi = dati[["data","stato","terapia_intensiva","totale_ospedalizzati","nuovi_deceduti"]]
datinuovi
```

```python
datinuovi["data"] = pd.to_datetime(datinuovi['data'])
datinuovi['data'] = datinuovi['data'].dt.date
datinuovi['data']= pd.to_datetime(datinuovi['data'])
```

```python
dati_MA7 = datinuovi
```

```python
dati_MA7.set_index('data', inplace=True)
```

```python
dati_MA7 = dati_MA7.rolling(window=7, min_periods=1).mean().round()
```

```python
dati_MA7.reset_index(inplace=True)
dati_MA7
```

```python
dati_MA7[["terapia_intensiva", "totale_ospedalizzati","nuovi_deceduti"]] = dati_MA7[["terapia_intensiva", "totale_ospedalizzati","nuovi_deceduti"]].astype(np.int64)
```
